```python
import math
import sys
import warnings; warnings.simplefilter('ignore')
import numpy as np
import random
import time
import os
import matplotlib.pyplot as plt
import seaborn as sns; sns.set(style='white')
import pandas as pd
from matplotlib import patches
from matplotlib.patches import Polygon
from matplotlib import pylab
from sklearn import decomposition
from sklearn import datasets
from sklearn.manifold import TSNE
from mpl_toolkits.mplot3d import Axes3D
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans, MeanShift
from sklearn.datasets import make_blobs
from sklearn.metrics import silhouette_score
from sklearn.manifold import TSNE
from sklearn.metrics import davies_bouldin_score
from sklearn.metrics import jaccard_score
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import DBSCAN
from sklearn.neighbors import NearestNeighbors
from sklearn.cluster import AffinityPropagation
from sklearn.datasets import make_blobs
from sklearn.cluster import SpectralClustering
from sklearn.cluster import AgglomerativeClustering
from sklearn.cluster import OPTICS
from sklearn.mixture import GaussianMixture
from sklearn.cluster import Birch
from sklearn.cluster import AgglomerativeClustering
from scipy.cluster.hierarchy import linkage, dendrogram
from pyspark.ml.clustering import BisectingKMeans
from pyspark.ml.evaluation import ClusteringEvaluator
import hdbscan

sys.path.append('D:\Diploma\cpu\openpose\examples\media')

dir_path = "D:\Diploma\cpu\openpose\examples\media"
image_files = [f for f in os.listdir(dir_path) if f.endswith('.jpg.pose')]
image_files += [f for f in os.listdir(dir_path) if f.endswith('.png.pose')]
print(image_files)
num_of_points = 25
num_of_objects = 10
standardization = True
colors = ['r', 'g', 'b', 'c', 'm', 'y', 'k']

def parse_coordinates(filename):
    with open(filename, "r") as file:
        lines = file.readlines()

    points = []

    for line in lines:
        x, y = line.split()

        x = float(x)
        y = float(y)
        points.append([x, y])

    return points


class AnalysisVectors:
    def __init__(self, clusterization):
        self.vectors = [] 
        for object in clusterization.clust_objects:
            self.vectors.append(object.points) 
        self.i_th_points = [[[self.vectors[j][i] for j in range(len(self.vectors))] for i in range(len(self.vectors[0]))]]
        
    def i_th_table_analys(self, i_th):   
        print(f"____________________________________________") 
        print(f"Точка {i_th}") 
        tablOfPoint = pd.DataFrame(self.i_th_points[0][i_th]) 
        print(tablOfPoint)
        variances = tablOfPoint.var()
        print(f"Дисперсия\txy:{variances}")
        
    def i_th_vector_analys(self, i_th):    
        x = np.array(self.i_th_points[0][i_th])[:, 0]
        print("Медиана\t\tx:", np.median(x))
        print("Матожидание\tx:", np.mean(x))
        print("Дисперсия\tх:", np.var(x))
        y = np.array(self.i_th_points[0][i_th])[:, 1]
        print("Медиана\t\ty:", np.median(y))
        print("Матожидание\ty:", np.mean(y))
        print("Дисперсия\ty:", np.var(y))

    # Данный анализ показывает, что никакой зависимости между координатами x и y нет, тем более в пределах 1 скелета (к менее 0,8 не считаются значимыми, а по диагонали)
    def i_j_th_vectors_analys(self, i_th, j_th):
        print(f"____________________________________________") 
        xi = np.array(self.i_th_points[0][i_th])[:, 0]
        xj = np.array(self.i_th_points[0][j_th])[:, 0]
        x = []
        x.append(xi)
        x.append(xj)
        yi = np.array(self.i_th_points[0][i_th])[:, 1]
        yj = np.array(self.i_th_points[0][j_th])[:, 1]
        y = []
        y.append(yi)
        y.append(yj)

        print(f"********************************************") 
        print(x)
        print("Коэффициент корреляции Пирсона между x ", i_th," и ",j_th,": \n", np.corrcoef(x, rowvar=False))
        print(y)
        print("Коэффициент корреляции Пирсона y ", i_th," и ",j_th,": \n", np.corrcoef(y, rowvar=False))

    def i_th_object_analys(self, i_th):
        print(f"____________________________________________") 
        print("Координаты ", i_th, "-го скелета: \n",self.vectors[i_th])
        print(f"********************************************") 
        print("Коэффициент корреляции Пирсона: \n", np.corrcoef(self.vectors[i_th], rowvar=False))


    def plot_i_th_points(self, i_th, points = None):
        print("\n", i_th,"-е точки:")
        x = [point[0] for point in self.i_th_points[0][i_th]]
        y = [point[1] for point in self.i_th_points[0][i_th]]
        plt.scatter(x, y)
        
        plt.show()

    def plot_all_points(self, points = None):
        
        print("\nВсе точки на графике:")
        
        for i in range(num_of_points):
            x = [point[0] for point in self.i_th_points[0][i]]
            y = [point[1] for point in self.i_th_points[0][i]]
            
            plt.scatter(x, y, label=i)
        
        plt.legend()
        plt.xlim(-0.1, 1)
        plt.show()       



class ClusterizationObject:
    def __init__(self, points, class_id, file_name):
        self.points = points
        self.class_id = class_id
        self.file_name = file_name
        

class Clusterization:
    def __init__(self, clust_objects: ClusterizationObject):
        self.clust_objects = clust_objects
        self.objects_id_in_clusters = []
        
        self.matrix_of_all_skeletons = [] 
        for object in self.clust_objects:
            self.matrix_of_all_skeletons.append(object.points) 
            
        self.matrix_of_all_skeletons_alternate  = []
        for sub_list in self.matrix_of_all_skeletons:
            flat_sub_list = []
            for item in sub_list:
                if isinstance(item, list):
                    flat_sub_list.extend(item)
                else:
                    flat_sub_list.append(item)
            self.matrix_of_all_skeletons_alternate.append(flat_sub_list)
        
    @staticmethod
    def normalize_points(points):            
        chest_number = 1 # номер точки – центра координат  (груди)

        relative_coords = [[0.0, 0.0]] * num_of_points
        max_axial_deviation = 0
        for i in range(len(points)):
            
            # (1) Рассчет относительных координат
            relative_coords[i] = [points[i][0] - points[chest_number][0], points[i][1] - points[chest_number][1]]
            
            # (2) Поиск максимальной координаты (max[xi;yi])
            max_axial_deviation = max(max_axial_deviation, abs(relative_coords[i][0]), abs(relative_coords[i][1]))
       
        normalization_koeff = 2 * max_axial_deviation
        
        # (3) Нормализация координат максимальной координатой
        for i in range(len(points)):
            relative_coords[i] = [relative_coords[i][0] / normalization_koeff, relative_coords[i][1] / normalization_koeff]

        new_points = [[0.0, 0.0]] * num_of_points
        
        # (4) Вычисление новых относительных координат с координатой груди в (0.5;0.5)
        for i in range(len(points)):
            new_points[i] = [0.5 + relative_coords[i][0], 0.5 + relative_coords[i][1]]

        return new_points

    @staticmethod
    def sum_points(a, b):
        c = []
        for i in range(min(len(a), len(b))):
            c.append([a[i][0] + b[i][0], a[i][1] + b[i][1]])

        if len(a) > len(b):
            c += a[len(b):]
        elif len(b) > len(a):
            c += b[len(a):]

        return c


    @staticmethod
    def average_points(point_lists):
        averages = []
        for i in range(num_of_points):
            x_sum = 0
            y_sum = 0
            for point_list in point_lists:
                x_sum += point_list[i][0]
                y_sum += point_list[i][1]
            x_avg = x_sum / len(point_lists)
            y_avg = y_sum / len(point_lists)
            averages.append((x_avg, y_avg))
        return averages

    
    def clustering(self, func):
        def wrapper(*args, **kwargs):
            
            print("-------------------------------------------------------")
            print("\nМЕТОД: ", func,"\n")
            print("-------------------------------------------------------")
            print("Без уменьшения размерности")
            
            clustering = func(*args, **kwargs)
            
            t0 = time.time()
            
            clustering.fit(self.matrix_of_all_skeletons_alternate)
            
            print("Затрачено времени:\t", time.time() -  t0)
            print(clustering.labels_)
            print("Оцекнка качества:")
            print("Коэффициент силуэта:\t", silhouette_score(self.matrix_of_all_skeletons_alternate, clustering.labels_)) # от -1 до 1, 1 - супер, -1 - фигня
            print("Индекс Дэвиса-Болдуина:\t", davies_bouldin_score(self.matrix_of_all_skeletons_alternate, clustering.labels_))
           
            for i, obj in enumerate(self.clust_objects):
                obj.class_id = clustering.labels_[i]
                
            t0 = time.time()  
            
            pca = PCA(n_components=2)
            X_pca = pca.fit_transform(np.array(self.matrix_of_all_skeletons_alternate))
            
            print("\nЗатрачено времени на нормализацию:\t", time.time() -  t0)
            
            print("\nРазмерность уменьшена до 2")
            
            t0 = time.time()
            
            clustering.fit(X_pca)
            
            print("Затрачено времени:\t", time.time() -  t0)
            
            print(clustering.labels_)
            print("Оцекнка качества:")
            print("Коэффициент силуэта:\t", silhouette_score(self.matrix_of_all_skeletons_alternate, clustering.labels_)) # от -1 до 1, 1 - супер, -1 - фигня
            print("Индекс Дэвиса-Болдуина:\t", davies_bouldin_score(self.matrix_of_all_skeletons_alternate, clustering.labels_))# от 0 до +оо, чем меньше, тем лучше
            
            for i in range(len(X_pca)):
                plt.scatter(X_pca[i, 0], X_pca[i, 1], color=colors[clustering.labels_[i]], alpha=0.5)
            plt.show()
        
        return wrapper    
    
    def clustering_probabilistic(self, func):
        def wrapper(*args, **kwargs):
            
            print("-------------------------------------------------------")
            print("\nМЕТОД: ", func,"\n")
            print("-------------------------------------------------------")
            print("Без уменьшения размерности")
            
            
            clustering = func(*args, **kwargs)
            
            t0 = time.time()
            
            clustering.fit(self.matrix_of_all_skeletons_alternate)
            
            labels = clustering.predict(self.matrix_of_all_skeletons_alternate)
            print("Затрачено времени:\t", time.time() -  t0)
            print(labels)
            print("Оцекнка качества:")
            print("Коэффициент силуэта:\t", silhouette_score(self.matrix_of_all_skeletons_alternate, labels)) # от -1 до 1, 1 - супер, -1 - фигня
            print("Индекс Дэвиса-Болдуина:\t", davies_bouldin_score(self.matrix_of_all_skeletons_alternate, labels))
           
            for i, obj in enumerate(self.clust_objects):
                obj.class_id = labels[i]
            
            t0 = time.time()  
            
            pca = PCA(n_components=2)
            X_pca = pca.fit_transform(np.array(self.matrix_of_all_skeletons_alternate))
            
            print("\nЗатрачено времени на нормализацию:\t", time.time() -  t0)
            
            print("\nРазмерность уменьшена до 2")
            
            t0 = time.time()
            
            clustering.fit(X_pca)
            labels = clustering.predict(X_pca)
            
            print("Затрачено времени:\t", time.time() -  t0)
            
            print(labels)
            print("Оцекнка качества:")
            print("Коэффициент силуэта:\t", silhouette_score(self.matrix_of_all_skeletons_alternate, labels)) # от -1 до 1, 1 - супер, -1 - фигня
            print("Индекс Дэвиса-Болдуина:\t", davies_bouldin_score(self.matrix_of_all_skeletons_alternate, labels))# от 0 до +оо, чем меньше, тем лучше
            
            for i in range(len(X_pca)):
                plt.scatter(X_pca[i, 0], X_pca[i, 1], color=colors[labels[i]], alpha=0.5)
            plt.show()
        
        return wrapper
              
    def ward_hierarchical_clustering(self):
        t0 = time.time()
        Z = linkage(self.matrix_of_all_skeletons_alternate, 'ward')
        plt.figure(figsize=(10, 5))
        plt.title("Ward Dendrogram")
        dendrogram(Z)
        print("Затрачено времени:\t", time.time() -  t0)
        plt.show()
    
if __name__ == "__main__":
    clustObjects = []
    for current_index in range(len(image_files)):
        file_path = os.path.join(dir_path, image_files[current_index])
        points = parse_coordinates(file_path)
        normalized_points = Clusterization.normalize_points(points)
        clustObjects.append(ClusterizationObject(normalized_points, None, image_files[current_index]))
    clusterization = Clusterization(clustObjects)

    clusterization.clustering(MeanShift)()
    clusterization.clustering(KMeans)(init="k-means++", n_clusters=3, n_init=10)
    clusterization.clustering(DBSCAN)(eps=0.5, min_samples=2)
    #clusterization.clustering(hdbscan.HDBSCAN)(min_cluster_size=4, metric='euclidean')
    clusterization.clustering(SpectralClustering)(n_clusters=3, affinity='nearest_neighbors',n_neighbors=3)
    clusterization.clustering(AgglomerativeClustering)(n_clusters=3, linkage='ward')
    clusterization.clustering(OPTICS)(min_samples=2, xi=.1)
    #clusterization.clustering_probabilistic(BisectingKMeans)(k=3, seed=1)
    clusterization.clustering_probabilistic(Birch)(threshold=0.5, n_clusters=None, branching_factor=50, compute_labels=True, copy=True)
    clusterization.clustering_probabilistic(GaussianMixture)(n_components=3)
    clusterization.ward_hierarchical_clustering()
    
    points = np.array([clustObject.points for clustObject in clustObjects])


    #ANALYSIS _____________________________________________________________
    analysisVectors = AnalysisVectors(clusterization)
    
    for i in range(25):
       analysisVectors.plot_i_th_points(i)
    analysisVectors.plot_all_points() 
    
    for i in range(25):
        analysisVectors.i_th_table_analys(i) 
        analysisVectors.i_th_vector_analys(i) 
        
    for i in range(num_of_objects):
        analysisVectors.i_th_object_analys(i)
        
    for i in range(25):
        analysisVectors.i_j_th_vectors_analys(24,i)
    print(clusterization.clust_objects[0].points)
```
